```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

import os
```

```python
data_url = "https://github.com/mwaugh0328/data_bootcamp_spring_2019/raw/master/intro_to_matplotlib/clean/usgdp_data.xlsx"
```

```python
data_url
```

```python
usdata = pd.read_excel(data_url, index_col = 0)
```

```python
usdata.head()
```

```python
usdata.index
```

```python
usdata.shape
```

```python
usdata["2015":"2016"]
```

```python
usdata["gdp_growth"] = 100*4*usdata["gdp"].pct_change()

usdata["cons_growth"] = 100*4*usdata["cons"].pct_change()
```

```python
usdata.head()
```

```python
usdata.columns
```

```python
var_list = ["gdp", "cons", "invs", "gov", "nx"]

for var in var_list:
    
    new_col_name = var + "_growth"
    
    usdata[new_col_name] = 100*4*usdata[var].pct_change()
```

```python
fig, ax = plt.subplots()

ax.plot(usdata.index, usdata.gdp_growth)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(usdata.index, usdata.gdp_growth, color = "b", linewidth = 3)

ax.set_title("GDP Growth", fontsize = 14, fontweight = "bold")
ax.set_ylabel("Annualized Growth Rate")
ax.set_xlabel("Year")

ax.set_ylim(-10,10)
ax.set_xlim(dt.datetime(2005,1,1),dt.datetime(2018,10,1)) 

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

plt.savefig("us_gdp_growth.png", bbox_inches = "tight", dip = 1200)

plt.show()
```

```python
def my_plot_fun(var):
    
    growth_var = var + "_growth"
    
    fig, ax = plt.subplots()

    ax.plot(usdata.index, usdata[growth_var], color = "b", linewidth = 3)

    ax.set_title(var.capitalize() + " Growth", fontsize = 14, fontweight = "bold")
    ax.set_ylabel("Annualized Growth Rate")
    ax.set_xlabel("Year")

    ax.set_ylim(usdata[growth_var].min() - 1,usdata[growth_var].max() + 1)
    ax.set_xlim(dt.datetime(2005,1,1),dt.datetime(2018,10,1)) 

    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)

    #plt.savefig("us_gdp_growth.png", bbox_inches = "tight", dip = 1200)

    plt.show()
    
```

```python
my_plot_fun("nx")
```

```python
fig, ax = plt.subplots()

ax.plot(usdata.index, usdata.gdp_growth, color = "b", linewidth = 3, 
        alpha = 0.25, label = "GDP")

ax.plot(usdata.index, usdata.cons_growth, color = "r", linewidth = 3,
        linestyle = '--', label = "Consumption")

ax.plot(usdata.index, usdata.invs_growth, color = "k", linewidth = 3,
        linestyle = '--', label = "Investment")

ax.legend(frameon = False)

#ax.set_title("GDP Growth", fontsize = 14, fontweight = "bold")
ax.set_ylabel("Annualized Growth Rate")
ax.set_xlabel("Year")

#ax.set_ylim(-10,10)
ax.set_xlim(dt.datetime(2005,1,1),dt.datetime(2018,10,1)) 

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

#plt.savefig("us_gdp_growth.png", bbox_inches = "tight", dip = 1200)

plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, sharex = True, figsize = (12,4))

ax[0].plot(usdata.index, usdata.gdp_growth, color = "b", linewidth = 3, 
        alpha = 0.5, label = "GDP")

ax[1].plot(usdata.index, usdata.cons_growth, color = "r", linewidth = 3, 
        alpha = 0.5, label = "Consumption")

ax[0].set_title("GDP")

ax[0].set_ylabel("Growth Rate, Percent")

fig.suptitle("GDP and Consumption Growth")



plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(usdata.gdp_growth, usdata.cons_growth, alpha = 0.50)

ax.scatter(usdata.gdp_growth.sort_values()[0:5],
           usdata.cons_growth.loc[usdata.gdp_growth.sort_values()[0:5].index],
                                                           c = 'r', alpha = 0.95, s = 50)

ax.set_ylim(-10,10)
ax.set_xlim(-10,10)

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.set_title("GDP vs Consumpiton")

############################################################################
# This is something new I learned recently

my_message = "What Year is This? "

my_message = my_message + usdata.gdp_growth.idxmin().strftime('%m/%d/%Y')

ax.annotate(
    my_message, 
    xy=(-8.25, -3.75), # This is where we point at...
    xycoords="data", # Not exactly sure about this
    xytext=(0, -9), # This is about where the text is
    horizontalalignment="left", # How the text is alined
    arrowprops={
        "arrowstyle": "-|>", # This is stuff about the arrow
        "connectionstyle": "angle3,angleA=5,angleB=0",
        "color": "black"
    },
    fontsize=12,
)

############################################################################

plt.show()
```
